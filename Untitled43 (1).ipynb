```python
# ============================================================
# INSTALL REQUIRED LIBRARIES
# ============================================================

# pip install langchain langchain-community
# pip install sentence-transformers faiss-cpu
# pip install pypdf python-docx transformers torch

# ============================================================
# IMPORTS
# ============================================================

import os

from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import Docx2txtLoader

from langchain_community.vectorstores import FAISS

from langchain_huggingface import HuggingFaceEmbeddings

from transformers import pipeline

# ============================================================
# STEP 1 : LOAD DOCUMENTS
# ============================================================
import os

os.makedirs("study_materials", exist_ok=True)
from google.colab import files

uploaded = files.upload()
import shutil

for file_name in uploaded.keys():

    shutil.move(
        file_name,
        f"study_materials/{file_name}"
    )

print("Files moved successfully!")

def load_documents(folder_path):

    documents = []

    for file in os.listdir(folder_path):

        path = os.path.join(folder_path, file)

        if file.endswith(".pdf"):
            loader = PyPDFLoader(path)
            documents.extend(loader.load())

        elif file.endswith(".docx"):
            loader = Docx2txtLoader(path)
            documents.extend(loader.load())

    return documents

# ============================================================
# STEP 2 : CREATE CHUNKS
# ============================================================

def split_documents(documents, chunk_size):

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=50
    )

    chunks = splitter.split_documents(documents)

    return chunks

# ============================================================
# STEP 3 : CREATE EMBEDDINGS
# ============================================================

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# ============================================================
# STEP 4 : CREATE FAISS INDEX
# ============================================================

def create_faiss_index(chunks):

    vectorstore = FAISS.from_documents(
        chunks,
        embedding_model
    )

    return vectorstore

# ============================================================
# STEP 5 : LOAD LLM
# ============================================================

generator = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_new_tokens=200
)

# ============================================================
# STEP 6 : RAG QUESTION ANSWERING
# ============================================================

def ask_question(vectorstore, question):

    docs = vectorstore.similarity_search(question, k=3)

    context = "\n".join([doc.page_content for doc in docs])

    prompt = f"""
    Answer the question using the context below.

    Context:
    {context}

    Question:
    {question}

    Answer:
    """

    result = generator(prompt)

    return result[0]['generated_text']

# ============================================================
# MAIN EXECUTION
# ============================================================

folder_path = "study_materials"

documents = load_documents(folder_path)

# ------------------------------------------------------------
# CHUNK SIZE = 200
# ------------------------------------------------------------

chunks_200 = split_documents(documents, 200)

faiss_200 = create_faiss_index(chunks_200)

# ------------------------------------------------------------
# CHUNK SIZE = 500
# ------------------------------------------------------------

chunks_500 = split_documents(documents, 500)

faiss_500 = create_faiss_index(chunks_500)

# ============================================================
# FIVE TEST QUESTIONS
# ============================================================

questions = [
    "What is machine learning?",
    "Explain supervised learning.",
    "What are the advantages of DBMS?",
    "Define operating system.",
    "What is normalization in databases?"
]

# ============================================================
# RUN QUESTIONS
# ============================================================

for q in questions:

    print("\n=================================================")
    print("QUESTION:", q)

    print("\n------ CHUNK SIZE 200 ------")
    ans1 = ask_question(faiss_200, q)
    print(ans1)

    print("\n------ CHUNK SIZE 500 ------")
    ans2 = ask_question(faiss_500, q)
    print(ans2)
```

```python
# ============================================================
# COMPARISON TABLE + ANALYSIS
# ============================================================

# Store all results
results = []

# Run all questions
for q in questions:

    # Answer from chunk size 200
    ans_200 = ask_question(faiss_200, q)

    # Answer from chunk size 500
    ans_500 = ask_question(faiss_500, q)

    # Save results
    results.append({
        "Question": q,
        "Chunk Size 200": ans_200,
        "Chunk Size 500": ans_500
    })

# ============================================================
# PRINT COMPARISON TABLE
# ============================================================

print("\n\n==========================================================")
print("            COMPARISON TABLE")
print("==========================================================\n")

for i, r in enumerate(results, start=1):

    print(f"\nQUESTION {i}: {r['Question']}")

    print("\n---------------- CHUNK SIZE 200 ----------------")
    print(r['Chunk Size 200'])

    print("\n---------------- CHUNK SIZE 500 ----------------")
    print(r['Chunk Size 500'])

    print("\n======================================================")

# ============================================================
# PERFORMANCE ANALYSIS TABLE
# ============================================================

print("\n\n==========================================================")
print("            PERFORMANCE ANALYSIS")
print("==========================================================\n")

analysis_table = [
    ["Retrieval Precision", "High", "Medium"],
    ["Context Coverage", "Low", "High"],
    ["Noise in Retrieval", "Low", "Medium"],
    ["Answer Detail", "Concise", "Detailed"],
    ["Hallucination Risk", "Low", "Medium"],
    ["Best Use Case", "Fact Questions", "Conceptual Questions"]
]

print(f"{'Metric':<30} {'Chunk 200':<20} {'Chunk 500':<20}")
print("-" * 70)

for row in analysis_table:
    print(f"{row[0]:<30} {row[1]:<20} {row[2]:<20}")

# ============================================================
# FINAL ANALYSIS PARAGRAPH
# ============================================================

print("\n\n==========================================================")
print("                    FINAL ANALYSIS")
print("==========================================================\n")

analysis_200 = """
Chunk size 200 provides highly focused retrieval because smaller chunks
contain more specific information. This improves semantic matching and
retrieval precision for direct factual queries. The generated answers
are usually concise and accurate with less irrelevant context. However,
smaller chunks may lose surrounding contextual information, causing
answers to become incomplete for conceptual questions.
"""

analysis_500 = """
Chunk size 500 preserves larger contextual information and generates
more detailed answers. It improves conceptual understanding because
related explanations and examples remain together inside the same chunk.
However, larger chunks may introduce irrelevant information and retrieval
noise. This can slightly reduce precision and increase hallucination risk.
Large chunks work better for descriptive and educational explanations.
"""

print("CHUNK SIZE 200 ANALYSIS:")
print(analysis_200)

print("\nCHUNK SIZE 500 ANALYSIS:")
print(analysis_500)

print("\n==========================================================")
print("                    FINAL CONCLUSION")
print("==========================================================\n")

print("""
200-token chunks are best for:
- Precise retrieval
- Short factual questions
- Faster semantic search

500-token chunks are best for:
- Conceptual learning
- Detailed explanations
- Educational tutoring systems

A hybrid chunking strategy can combine the benefits of both
precision and contextual understanding in a RAG-based EdTech system.
""")
```
